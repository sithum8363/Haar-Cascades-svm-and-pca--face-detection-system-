```python
from sklearn.svm import SVC 
from sklearn.decomposition import PCA 
from sklearn.model_selection import train_test_split
import  numpy as nu
data=nu.load('data.npy')
target=nu.load('target.npy')
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data,target, test_size=0.2,random_state=42)

pca=PCA(n_components=50,whiten=True)
X_train_pca=pca.fit_transform(X_train)
X_test_pca=pca.transform(X_test)
```

```python
svc = SVC(kernel='rbf', C=15, gamma=0.001)
svc.fit(X_train_pca, y_train)
pred = svc.predict(X_test_pca)
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(y_test,pred)
print(acc)
```

```python
predicted_target=svc.predict(X_test_pca)
target=nu.unique(target)
print(target)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predicted_target,target_names=['0','1','2','3','4']))
```

```python
import joblib
joblib.dump(svc,'SVM-Face Recognition.sav')
joblib.dump(pca,'pca-Face Recognition.sav')
```
