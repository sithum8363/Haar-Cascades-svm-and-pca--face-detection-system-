```python
import cv2
import collections
import os
import  numpy as nu
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

p=r'tainin data'
labels=os.listdir(p)
categories=nu.arange(len(labels))
category_dict=dict(zip(labels,categories))
nu.save('category_dict',category_dict)
data=[]
target=[]

for label in labels:
    
    imgs_path=os.path.join(p,label)
    img_names=os.listdir(imgs_path)
    
    for img_name in img_names:
        
        img_path=os.path.join(imgs_path,img_name)
        img=cv2.imread(img_path)
    
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_classifier.detectMultiScale(gray)
        
        try:
        
            for (x,y,w,h) in faces:

                cropped_face=gray[y:y+h,x:x+w]
                cv2.imshow('cropped_face',cropped_face)

                k=cv2.waitKey(0)

                if(k==121):
                    cropped_face=cv2.resize(cropped_face,(50,50))
                    data.append(cropped_face)
                    target.append(category_dict[label])
                    print("value in ")
                else:
                    print("value out ")
                    pass       
        except:
           
            pass
cv2.destroyAllWindows()
data=nu.array(data)
data=data.reshape(data.shape[0],data.shape[1]*data.shape[2])
target=nu.array(target)
print(data[:5])
print(target[:5])
nu.save('data',data)
nu.save('target',target)

print(collections.Counter(target))
```
